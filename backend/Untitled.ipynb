```python
import pandas as pd
import geopandas as gpd
from shapely.geometry import Point
from geopy.distance import great_circle
import json

def load_trajectories(file_path):
    """Load DDR/STATFOR CSV → clean GeoDataFrame"""
    df = pd.read_csv(file_path)  # Expects: flight_id, lat, lon, alt_ft, time_min, od_pair
    gdf = gpd.GeoDataFrame(
        df, 
        geometry=[Point(lon, lat) for lon, lat in zip(df.lon, df.lat)]
    )
    # Quick GC baseline
    gdf['gc_nm'] = 0.0  # Placeholder for now
    return gdf

# Test with sample (replace with your DDR)
if __name__ == "__main__":
    # Create sample data if no file
    sample = {
        'flight_id': ['UA123', 'UA123', 'PL456', 'PL456'],
        'lat': [50.0, 50.5, 49.8, 50.2],
        'lon': [30.0, 30.5, 29.8, 30.2],
        'alt_ft': [35000, 35500, 34000, 34500],
        'time_min': [0, 15, 0, 12],
        'od_pair': ['KBP-WAW', 'KBP-WAW', 'LWO-KRK', 'LWO-KRK']
    }
    df_sample = pd.DataFrame(sample)
    df_sample.to_csv('../data/ua_sample/sample_trajectories.csv', index=False)
    
    gdf = load_trajectories('../data/ua_sample/sample_trajectories.csv')
    print("Loaded trajectories:")
    print(gdf.head())
    gdf.to_file('../output/sample_trajectories.geojson', driver='GeoJSON')
    print("Saved to output/sample_trajectories.geojson")
```
